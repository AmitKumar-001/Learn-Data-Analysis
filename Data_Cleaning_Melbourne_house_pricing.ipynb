```python
#removing the columns whose are mostly missing values like more than 30%
df.drop(['BuildingArea', 'YearBuilt', 'CouncilArea'], axis=1, inplace=True) #the columns has removed
```

```python
df.shape #earliear it was 21 column now it is only 21
```

```python
#now no anyone has more than 30% missing values
round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#check if anyone has more than 5 missing values
df[df.isnull().sum(axis=1) > 5]
```

```python
#count the number of rows having > 5 missing values
# use len(df.index) #need to deep dive in it with help of another sources
len(df[df.isnull().sum(axis=1) > 5].index)
```

```python
#4278 rows have more than 5 missing values
#Calculate the percentage
round(100*(len(df[df.isnull().sum(axis=1) > 5].index)/len(df.index)), 2)
```

```python
#retaining the rows only having < 5 missing values
df = df[df.isnull().sum(axis=1) <= 5]
```

```python
#now check summary agian
round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#removing missing price rows
df = df[~np.isnan(df['Price'])] #need to deep dive in it with external sources
```

```python
#now view summary again
round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#focus more missing value like landsize
df['Landsize'].describe() #do deep dive in it
```

Here the minimum is 0, max is 433014, the mean is 558 and median (50%) is 440. There is a significant variation in the 25th and 75th percentile as well (176 to 651). Thus, imputing this with mean/median seems quite biased and so we should remove the missing value.

```python
#removing missig values from landsize
df = df[~np.isnan(df['Landsize'])]
```

```python
#now recheck % missing
round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#now going to impute the missing value rather than removing
#rows having lattitude and longitude missing
df[np.isnan(df['Lattitude'])]
```

```python
#summary of the rows having missing lattitude and longitude so if possible we can impute the misssig value
#don't use like >> df[np.isnan(df['Lattitude'])].describe()
df.loc[:,['Lattitude','Longtitude']].describe() #deep dive to .loc[:,]
```

Here, there is no big difference between (max, min) or (mean, median) or variation from 25th to 75th percentile. (not comparing between Lattitude and Longtitude). We can impute the missing values by mean value respectively.

```python
#imputing Lattitude and Longitude by mean values
df.loc[np.isnan(df['Lattitude']), ['Lattitude']] = df['Lattitude'].mean()
df.loc[np.isnan(df['Longtitude']), ['Longtitude']] = df['Longtitude'].mean()
#need to deep dive
```

```python
#now check missing percentage
round(100*(df.isnull().sum()/len(df.index)), 2)
```

Still Bathroom and Car has missing values so now let either impute or remove

```python
#focus on Bathroom and Car
df.loc[:,['Bathroom', 'Car']].describe()
```

These two are integer type variables, and thus have values 0, 1, 2, 5 etc. We can't impute the missing values by the mean or the median (like 1.53 bathrooms doesn't make sense!).
Thus, we need to impute them by the mode (the most common occurring value)

```python
#converting to type 'category'
df['Car'] = df['Car'].astype('category') #need to deep dive

#displaying frequencies of each category
df['Car'].value_counts()
```

here mode is 2 as it is most common (5606) so we will replace the missing value by 2

```python
#replace missing value by 2
df.loc[pd.isnull(df['Car']),['Car']] = 2
```

```python
#now check the missing percentage
round(100*(df.isnull().sum()/len(df.index)), 2)
```

now only bathroom has missing value so now let work on it either impute or remove

```python
#Similarly doing for Bathroom as above did for car
#converting to type 'category'
df['Bathroom'] = df['Bathroom'].astype('category') #need to deep dive

#displaying frequencies of each category
df['Bathroom'].value_counts()
```

```python
#here mode of bathroom is 1 (7517 the highest) so we can impute the missing value by 1
df.loc[pd.isnull(df['Bathroom']),['Bathroom']] = 2
```

```python
#now check the missing percentage
round(100*(df.isnull().sum()/len(df.index)), 2)
```

Now none of them are missing

```python
#now no missing values
#review the data shape after cleaning the data
df.shape
```
